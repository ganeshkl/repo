```python
import requests
from bs4 import BeautifulSoup
import csv

# Base URL for Stack Overflow Python tag questions
url = 'https://stackoverflow.com/questions/tagged/python'

# Function to scrape questions from a given page
def scrape_questions(page_url):
    try:
        # Send a request to fetch the page content
        response = requests.get(page_url)
        response.raise_for_status()  # Raise HTTPError for bad responses

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all question elements
        questions = soup.find_all('a', class_='s-link')  # Updated class name

        # Return the list of questions
        return questions

    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
        return []

# List to store all questions
all_questions = []

# Loop through multiple pages (e.g., first 5 pages)
page_number = 1
while True:
    # Construct the URL for the current page
    page_url = f'{url}?page={page_number}'

    # Scrape questions from the current page
    questions = scrape_questions(page_url)

    # If no questions are found, stop scraping
    if not questions:
        break

    # Add questions from this page to the list
    all_questions.extend(questions)

    # Move to the next page
    page_number += 1

# Check if we found any questions
if not all_questions:
    print("No questions were found.")
    exit()

# Specify the CSV file name
csv_file_name = 'stackoverflow_all_python_questions.csv'

# Write all questions to the CSV file
with open(csv_file_name, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Question Title'])  # Write header row

    # Extract each question title and write it to the CSV
    for question in all_questions:
        title = question.get_text().strip()
        writer.writerow([title])

print(f"All questions have been successfully extracted and saved to '{csv_file_name}'.")

# Optionally, print out the first few questions to the console
print("\nExtracted Questions:")
for question in all_questions[:10]:  # Print first 10 questions
    print(question.get_text().strip())
```
